```python
class financeiro:
  
  def __init__(self, nome):
    self.nome = nome

  def serie(self):
    ggh = yf.Ticker(self.nome+'.SA')
    ghg = ggh.history(period="max")
    hgg = pd.DataFrame(ghg['Close'])
   
    hgg.columns = [self.nome]
   
    hgg['data'] = hgg.index
    hgg = hgg.reset_index(drop=True)
    hgg['data'] = pd.to_datetime(hgg['data'], format = '%Y/%m/%d')
    hgg['data'] = hgg.data.dt.strftime("%d/%m/%Y")
    hgg = hgg.set_index('data')
    
    display(hgg)

financeiro('ABEV3').serie()
```
